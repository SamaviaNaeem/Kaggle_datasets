```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/laptop_prices[1].csv')
```

```python
print("first 5 rows of the datasets:")
print(df.head())
```

```python
print("\nBasic information about the dataset:")
print(df.info())
```

```python
print("\nSummary statistics for numerical columns:")
print(df.describe())
```

```python
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```python
print("\nNumber of unique Product:")
print(df['Product'].nunique())
```

```python
print("\nCompany counts:")
print(df['Company'].value_counts())
```

```python
print("nTop 5 most Company records:")
print(df.sort_values (by='Ram' , ascending= False).head())
```

```python
print("\nTop 5 least Company records:")
print(df.sort_values (by='Ram' , ascending= True).head())
```

```python
print("\nCompany with the highest Price_euros  greater than 150:")
expensive_Company = df[df['Price_euros'] >150]
print(expensive_Company)
```

```python
print("\nAverage the lowest Price_euros for each Weight:")
avg_Price_euros = df.groupby('Weight')['Price_euros'] .mean()
print(avg_Price_euros)
```

```python
import matplotlib.pyplot as plt
```

```python
avg_Price_euros.plot(kind='hist',title='Average Price_euros by Company')
plt.xlabel('Company')
plt.ylabel('Average Price_euros')
plt.show()
```

```python
print("\nAverage the lowest Price_euros for each Company:")
avg_Price_euros = df.groupby('Company')['Price_euros'] .mean()
print(avg_Price_euros)
```
