```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/Customer_Purchasing_Behaviors[1].csv')
```

```python
print("first 5 rows of the datasets:")
print(df.head())
```

```python
print("\nBasic information about the dataset:")
print(df.info())
```

```python
print("\nSummary statistics for numerical columns:")
print(df.describe())
```

```python
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```python
print("\nNumber of unique age:")
print(df['age'].nunique())
```

```python
print("\nregion counts:")
print(df['region'].value_counts())
```

```python
print("nTop 5 most purchase_amount records:")
print(df.sort_values (by='annual_income' , ascending= False).head())
```

```python
print("\nTop 5 least purchase_amount records:")
print(df.sort_values (by='annual_income' , ascending= True).head())
```

```python
print("\nuser_id with the highest purchase_amount greater than 150:")
expensive_user = df[df['purchase_amount'] >150]
print(expensive_user)
```

```python
print("\nAverage the lowest purchase_amount for each user_id:")
avg_purchase_amount = df.groupby('user_id')['purchase_amount'] .mean()
print(avg_purchase_amount)
```

```python
import matplotlib.pyplot as plt

```

```python
avg_purchase_amount.plot(kind='hist',title='Average purchase_amount by purchase_frequency')
plt.xlabel('purchase_frequency')
plt.ylabel('Average purchase_amount')
plt.show()
```
