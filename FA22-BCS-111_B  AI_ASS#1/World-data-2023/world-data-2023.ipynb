```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/world-data-2023[1].csv')
```

```python
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```python
print("\nNumber of unique Country:")
print(df['Country'].nunique())
```

```python
print("nTop 5 most Country records:")
print(df.sort_values (by='GDP' , ascending= False).head())
```

```python
print("\nAverage the lowest Latitude for each Country:")
avg_Latitude = df.groupby('Country')['Latitude'] .mean()
print(avg_Latitude)
```

```python
import matplotlib.pyplot as plt
avg_Latitude.plot(kind='hist',title='Average Latitude by Country')
plt.xlabel('Country')
plt.ylabel('Average Latitude')
plt.show()
```
