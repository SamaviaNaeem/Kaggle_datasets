```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/16k_Movies[1].csv')
```

```python
print("first 15 rows of the datasets:")
print(df.head())
```

```python
print("\nBasic information about the dataset:")
print(df.info())
```

```python
print("\nSummary statistics for numerical columns:")
print(df.describe())
```

```python
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```python
print("\nNumber of unique Title:")
print(df['Title'].nunique())
```

```python
print("\nTitle counts:")
print(df['Title'].value_counts())
```

```python
print("nTop 5 most Release records:")
print(df.sort_values (by='Title' , ascending= False).head())
```

```python
print("\nTitle with the highest Rating ")
expensive_Title = df['Rating']
print(expensive_Title)
```

```python
print("\nAverage the lowest Rating for each Title:")
avg_rating = df.groupby('Title')['Rating'].mean()
print(avg_rating)
```

```python
import matplotlib.pyplot as plt
```

```python
print("\nDistribution of Ratings:")
plt.figure(figsize=(8, 6))
plt.hist(df['Rating'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show
```
