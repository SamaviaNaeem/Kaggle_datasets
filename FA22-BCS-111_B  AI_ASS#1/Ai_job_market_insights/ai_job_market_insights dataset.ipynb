```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/ai_job_market_insights[1].csv')
```

```python
print("first 5 rows of the datasets:")
print(df.head())
```

```python
print("\nBasic information about the dataset:")
print(df.info())
```

```python
print("\nSummary statistics for numerical columns:")
print(df.describe())
```

```python
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```python
print("\nNumber of unique Job_Title:")
print(df['Job_Title'].nunique())
```

```python
print("\Industry counts:")
print(df['Industry'].value_counts())
```

```python
print("nTop 5 most Required_Skills records:")
print(df.sort_values (by='Industry' , ascending= False).head())
```

```python
print("\nTop 5 least Location records:")
print(df.sort_values (by='Industry' , ascending= True).head())
```

```python
print("\nJob_Title with the highest Salary_USD greater than 70000:")
highest_salary = df[df['Salary_USD'] >70000]
print(highest_salary)
```

```python
print("\nAverage the lowest Salary_USD for each Industry:")
avg_salary = df.groupby('Industry')['Salary_USD'].mean()
print(avg_salary)
```

```python
import matplotlib.pyplot as plt
```

```python

avg_salary.plot(kind='bar',title='Average salary by Company_Size')
plt.xlabel('Company_Size')
plt.ylabel('Average salary')
plt.show()
```
