```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/World_Happiness_Report_2024[1].csv')
```

```python
print("first 5 rows of the datasets:")
print(df.head())
```

```python
print("\nBasic information about the dataset:")
print(df.info())
```

```python
print("\nSummary statistics for numerical columns:")
print(df.describe())
```

```python
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```python
print("\nNumber of unique year:")
print(df['year'].nunique())
```

```python
print("\nCountry name counts:")
print(df['Country name'].value_counts())
```

```python
print("nTop 5 most Life Ladder records:")
print(df.sort_values (by='Country name' , ascending= False).head())
```

```python
print("\nTop 5 least year records:")
print(df.sort_values (by='Country name' , ascending= True).head())
```

```python
print("\nCountry name with the highest Life Ladder greater than 5:")
expensive_Country_name = df[df['Life Ladder'] >5]
print(expensive_Country_name)
```

```python
print("\nAverage the lowest year for each Country name:")
avg_year = df.groupby('Country name')['year'] .mean()
print(avg_year)
```
